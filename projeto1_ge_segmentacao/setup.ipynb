```python
import os
import numpy as np 
from PIL import Image, ImageChops
from IPython.display import display
```

```python
folder_path = r'KITTI_MOTS_DATA\data_tracking_image_2\training\image_02\0019'
instance_folder_path = r'KITTI_MOTS_DATA\instances\instances\0019'

start_index = 0
file_iter = iter(os.listdir(folder_path)[start_index:])
current_index = start_index
```

```python
# read next frame
fname = next(file_iter)

# Read and display image
img_path = os.path.join(folder_path, fname)
img = Image.open(img_path)
# display(img)

# Read and display image instance
instance_path = os.path.join(instance_folder_path, fname)
img_instance = Image.open(instance_path)
# display(img_instance)

assert img.size == img_instance.size

# Apply mask instance in image
new_img_array = np.array(img)
img_instance_array = np.array(img_instance)

for i in range(new_img_array.shape[0]):
    for j in range(new_img_array.shape[1]):
        if img_instance_array[i][j] == 0:
            new_img_array[i][j] = [0,0,0]

new_img = Image.fromarray(np.uint8(new_img_array)).convert('RGB')
display(new_img)
```
